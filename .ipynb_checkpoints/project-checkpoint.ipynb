```python
from pyspark import SparkContext
from pyspark.sql import SparkSession
sc = SparkContext.getOrCreate()
ss = SparkSession.builder.getOrCreate()
```

```python
import pandas as pd
```

```python
data = pd.read_csv('./chicago_taxi_trips_2016_06.csv')
data.head()
```

```python
data1 = data.drop(['taxi_id', 'pickup_census_tract', 'dropoff_census_tract', 'company'], axis = 1)
```

```python
# missing value imputation; clean trip_seconds, trip_end_time and fare...
data1.drop(data1.loc[data1.fare.isna() | data1.trip_seconds.isna() | data1.dropoff_community_area.isna() | data1.pickup_community_area.isna()].index, axis = 0, inplace = True)
data1
```

```python
data1.isna().sum()
```

```python
data1['trip_start_timestamp'] = pd.to_datetime(data1['trip_start_timestamp'])
data1['trip_end_timestamp'] = pd.to_datetime(data1['trip_end_timestamp'])
```

```python
# feature engineering with the datetime variable (can not be used in tree model)
data1['month'] = data1['trip_start_timestamp'].dt.month
data1['dayofweek'] = data1['trip_start_timestamp'].dt.dayofweek
data1['start_hour'] = data1['trip_start_timestamp'].dt.hour
data1['end_hour'] = data1['trip_end_timestamp'].dt.hour
```

```python
data1 = data1.reset_index()
```

```python
data1.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data1['payment_num'] = pd.DataFrame(le.fit(data1['payment_type']).transform(data1['payment_type']))
```

```python
data1.isna().sum()
```

```python
data1.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data1.drop(['index', 'payment_type','trip_start_timestamp','trip_end_timestamp', 'fare', 'tips', 'tolls', 'extras'], axis = 1), data1['fare'], test_size=0.33, random_state=42)
```

```python
X_train
```

```python
regr = RandomForestRegressor(min_samples_leaf=10, max_depth=50, n_estimators=30)
regr.fit(X_train.values, y_train.values)
```

```python
regr.score(X_train, y_train)
```

```python
regr.score(X_test, y_test)
```
